### scraping all team name

```python
teams=[]
for i in soup.find_all('span',class_="u-hide-phablet"):
    teams.append(i.text)
    
teams
```

```python
matches1=[]
for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    matches1.append(i.text)
    
matches1
```

```python
points1=[]
for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    points1.append(i.text)
    
points1
```

```python
ratings1=[]
for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    ratings1.append(i.text)
    
ratings1
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Teams':teams,'Matches':matches1,'Points':points1,'Ratings':ratings1})
df
```
